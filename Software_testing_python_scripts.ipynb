```python
def count_files_in_path(directory_path):
    file_count = 0
    for _, _, files in os.walk(directory_path):
        file_count += len(files)
    return file_count
```

```python
def unzip_last_downloaded_zip():
    import os
    import zipfile

    # Set the path to the Downloads directory
    downloads_dir = r'C:\Users\rupes\Downloads'

    c1=count_files_in_path(downloads_dir)
    # Get the list of files in the Downloads directory
    files = os.listdir(downloads_dir)

    # Filter the files to keep only ZIP files
    zip_files = [f for f in files if f.endswith('.zip')]

    if zip_files:
        # Sort the ZIP files by modified timestamp in descending order
        zip_files.sort(key=lambda x: os.path.getmtime(os.path.join(downloads_dir, x)), reverse=True)

        # Extract the contents of the last downloaded ZIP file
        zip_path = os.path.join(downloads_dir, zip_files[0])
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(downloads_dir)

        # Get the name of the extracted folder (assuming only one top-level folder is extracted)
        extracted_folder = os.path.splitext(zip_files[0])[0]
        if c1<count_files_in_path(downloads_dir):
            png_files = [f for f in os.listdir(downloads_dir) if f.endswith('.png')]

            if png_files:
                # print("PNG file(s) found in the extracted folder.")
                for png_file in png_files:
                    print(png_file)

                return "passed"
            else:
                print("No PNG file found in the extracted folder.")

    else:
        print("No ZIP file found in the Downloads directory.")
```
